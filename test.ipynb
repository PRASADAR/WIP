```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# define true parameters
alpha_true = 1.5
beta_true = 0.5

# define impact decay function
def decay_func(t, alpha, beta):
    return (1 + alpha * t) ** (-beta)

# define simulated data
n = 1000
t = np.linspace(0, 1, n)
J = np.ones(n)
x = J * decay_func(t, alpha_true, beta_true) + 0.1 * np.random.randn(n)

# define log-likelihood function
def log_likelihood(theta):
    alpha, beta = theta
    return -0.5 * np.sum((x - J * decay_func(t, alpha, beta)) ** 2)

# maximize log-likelihood function using BFGS
res = opt.minimize(lambda theta: -log_likelihood(theta), x0=[1, 1], method='BFGS')

# extract estimated parameters
alpha_est, beta_est = res.x

# plot true and estimated decay functions
plt.plot(t, decay_func(t, alpha_true, beta_true), label='True Decay Function')
plt.plot(t, decay_func(t, alpha_est, beta_est), label='Estimated Decay Function')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Decay Function')
plt.show()

print('True Parameters: alpha = {}, beta = {}'.format(alpha_true, beta_true))
print('Estimated Parameters: alpha = {}, beta = {}'.format(alpha_est, beta_est))
```

```python
import numpy as np
from scipy.stats import norm, gamma
from skopt import gp_minimize
from skopt.space import Real, Space

# define true parameters
alpha_true = 1.5
beta_true = 0.5

# define impact decay function
def decay_func(t, alpha, beta):
    return (1 + alpha * t) ** (-beta)

# define simulated data
n = 1000
t = np.linspace(0, 1, n)
J = np.ones(n)
x = J * decay_func(t, alpha_true, beta_true) + 0.1 * np.random.randn(n)

# define log-likelihood function
def log_likelihood(params):
    alpha, beta = params
    return -0.5 * np.sum((x - J * decay_func(t, alpha, beta)) ** 2)

# define prior distributions for alpha and beta
alpha_prior = norm(loc=1.0, scale=0.5)
beta_prior = gamma(a=2.0, scale=1.0)

# define objective function for Bayesian optimization
def objective(params):
    alpha, beta = params
    log_prior = alpha_prior.logpdf(alpha) + beta_prior.logpdf(beta)
    log_likelihood_val = log_likelihood(params)
    return -(log_prior + log_likelihood_val)

# define search space for Bayesian optimization
space = Space([Real(0.0, 10.0, name='alpha'),
               Real(0.0, 1.0, name='beta')])

# run Bayesian optimization to find best parameters
res = gp_minimize(objective, space, n_calls=100, random_state=0)

# extract best parameters
alpha_est, beta_est = res.x

# plot true and estimated decay functions
plt.plot(t, decay_func(t, alpha_true, beta_true), label='True Decay Function')
plt.plot(t, decay_func(t, alpha_est, beta_est), label='Estimated Decay Function')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Decay Function')
plt.show()

print('True Parameters: alpha = {}, beta = {}'.format(alpha_true, beta_true))
print('Estimated Parameters: alpha = {}, beta = {}'.format(alpha_est, beta_est))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the exponential decay function
def exp_decay(t, a, b):
    return np.exp(-b * t) + a

# Generate simulated data
#np.random.seed(42)
t = np.linspace(0, 10, 101)
a_true, b_true = 0.5, 1.2
n_trades = 5
I0_true = np.random.rand(n_trades) 
u_true = np.random.rand(n_trades) 
x_obs = np.zeros_like(t)
for i in range(n_trades):
    x_obs += I0_true[i] * np.exp(-b_true * (t - u_true[i]))

# Fit the model using curve_fit
p0 = (0, 0) # initial guess for the parameters
popt, pcov = curve_fit(exp_decay, t, x_obs, p0)

# Print the estimated parameters
a_hat, b_hat = popt
print('Estimated a:', a_hat)
print('Estimated b:', b_hat)

# Plot the results
plt.plot(t, x_obs, label='Observed data')
plt.plot(t, exp_decay(t, a_true, b_true), label='True model')
plt.plot(t, exp_decay(t, a_hat, b_hat), label='Estimated model')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price dislocation')
plt.show()
```
